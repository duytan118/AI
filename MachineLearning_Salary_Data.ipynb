```python
import pandas as pd
url = 'Salary_Data.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()#ve histogram
plt.show
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.pairplot(data)
```

```python
data.plot(kind = 'scatter',x='Salary',y='YearsExperience',figsize = (9,9),color='black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Salary']),y = data['YearsExperience'])#fit la ham de hoc
print(model.intercept_)#HE SO
print(model.coef_)#HE SO GOC, Y = COEF*X + INTERCEPT
model.score(X = pd.DataFrame(data['Salary']),y = data['YearsExperience'])#do chinh xac
```

```python
data.plot(kind = 'scatter',x='Salary',y='YearsExperience',figsize = (9,9),color='black')
y_pred = model.predict(X=pd.DataFrame(data['Salary']))
plt.plot(data['Salary'],y_pred, color='red')
```

```python
model = linear_model.LinearRegression()#tao model moi
X = pd.DataFrame([data['Salary'],data['Salary']**2]).T    #tuyen tinh bac 2,   .T de chuyen vi
model.fit(X,y=data['YearsExperience'])

print('Intercept cua mo hinh: ',model.intercept_)
print('Coeffient cua mo hinh: ',model.coef_)
print('Do chinh xac cua mo hinh: ',model.score(X,y=data['YearsExperience']))
```

```python
import numpy as np
range = np.arange(30000,130000,0.1)
data.plot(kind = 'scatter',x='Salary',y='YearsExperience',figsize = (9,9),color='black')
predictor = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=predictor)
plt.plot(range, y_pred, color='blue')
```
